```python
from nltk.corpus import stopwords
```

```python
common_storp_words=stopwords.words('english')
```

```python
custom_stop_words=['amp', 'rt']
```

```python
import numpy as np
```

```python
tokens.value_counts()
```

```python
stop_words=np.hstack([common_storp_words,custom_stop_words])
```

```python
tokens_freq = pd.Series(tokens).value_counts()
```

```python
df_tokens=pd.DataFrame(tokens_freq).reset_index().rename(columns={'index':'token',0:'frequency'})
```

```python
df_tokens
```

```python
df_tokens=df_tokens[~df_tokens['token'].isin(stop_words)]
```
