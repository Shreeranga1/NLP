## Clean and Pre-process the data. 

```python
import pandas as pd
df=pd.read_csv('data_set.csv')
df.head()
```

```python
txt=list(df['review_text'])
```

```python
from nltk.tokenize import RegexpTokenizer
tokens=[]
for i in txt:
    tokens.append(RegexpTokenizer('\w+').tokenize(i))
tokens
```

```python
low=[]
for i in tokens:
    for j in i:
        low.append(j.lower())
low
```

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
tokens_final=[]
for i in low:
    if i not in stop_words:
        tokens_final.append(i)
tokens_final        
```

# Question : 2.(Marks : 15)

##	Perform a Sentiment analysis to understand the emotion of the individual reviews as well as the     overall review. 

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer=SentimentIntensityAnalyzer()
```

```python
df['Score']=df['review_text'].apply(lambda s:analyzer.polarity_scores(s)['compound'])
df.head()
```

```python
senti=[]
for i in df['Score']:
    if i>0:
        senti.append('Positive')
    else:
        senti.append('Negative')
df['Sentiment']=senti
df.head()
```

```python
import matplotlib.pyplot as plt
plt.hist(df.Score)
```

```python
df.Sentiment.value_counts()
```

```python
df.Score.mean()
```

The overall score is 0.5 which shows that the product has POSITIVE review overall and the count of negative review is only 27 compared to 102 positive reviews.

## Find the TF & IDF  of the words 'Excellent' & 'order'.Interpret the results

```python
def word_tf(word,token):
    return float(token.count(word))/len(token)
```

```python
print(word_tf('excellent',tokens_final))
print(word_tf('order',tokens_final))
```

 Since 'Excellent' starts with a capital letter, the count of that word is zero as all capital letters were converted during pre processing so i am using 'excellent' with all lower case letters.

```python
import numpy as np
vocabulary=set(tokens_final)
word_index={w:index for index,w in enumerate(vocabulary)}
voc_size=len(vocabulary)
doc_count=len(txt)
word_doc_count=np.zeros(voc_size)

for word in tokens_final:
    indexes=[word_index[word]]
    word_doc_count[indexes]+=1.0
```

```python
word_idf=np.log(doc_count/(1+word_doc_count).astype(float))
print(word_idf[word_index['excellent']])
print(word_idf[word_index['order']])
```

For 'Excellent' TF if low and IDF is high compared to the word 'order' so it signifies that the word 'order' appears more in the vocabulary than the word 'Excellent'.

```python
def tf_idf(word,token):
    if word not in word_index:
        return 0
    return word_tf(word,token)*word_idf[word_index[word]]
print(tf_idf('excellent',tokens_final))
print(tf_idf('order',tokens_final))
```
